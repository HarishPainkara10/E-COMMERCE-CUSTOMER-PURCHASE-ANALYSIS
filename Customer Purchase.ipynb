```python
# BUILDING MODEL 
from sklearn.cluster import KMeans
X=Data.iloc[:,2:]
```

```python
# USING ELBOW METHOS TO FIND THE VALUE OF k
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

# PLOTTING INERTIA FOR DIFFERENT k VALUES
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Graph')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Within The Cluster Sum Of Square')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

# ASSUMING 'X' IS YOUR INPUT DATA
for k in range(2, 11):  # ITERATE OVER RANGE OF k (from 2 to 10)
    kmeans = KMeans(n_clusters=k, init='k-means++', max_iter=300, n_init=10, random_state=42)
    
# PERFORM KMEANS CLUSTERING AND GET THE CLUSTER LABELS
    kmeans_labels = kmeans.fit_predict(X)
    
# CALCULATE THE SILHOUETTE SCORE THE CURRENT CLUSTERING
    silhouette_avg = silhouette_score(X, kmeans_labels,metric='euclidean')
    
#PRINT THE SILHOUETTE SCORE FOR CURRENT VALUE OF k
    print(f'For k={k} --> Silhouette Score: {silhouette_avg:.3f}')

# AS WE SEE THE SILHOUETEE SCORE IS GREATER IN K=2 WHICH MEANS THERE WILL BE 2 CLUSTERS
```

```python
kmeans_labels=KMeans(n_clusters=2,init='k-means++',max_iter=200,random_state=42)
```

```python
Data['Cluster']=kmeans_labels.fit_predict(X)
```

```python
Cluster_Total=Data.groupby('Cluster').describe()
Cluster_Total
```

```python
plt.figure(figsize=(15,6))
sns.scatterplot(x='Cluster',y='Orders',hue='Cluster',data=Data)
```

```python
# THIS SEGMENTATION MODEL HAS TWO DISTINCT CUSTOMER SEGMENTS:-

# Low_Volume_Customer:
  # This group consists people with smaller orders.

# High_Volume_Customer:
  # This group consists people with larger orders.

Cluster_Total={0:'Low_Volume_Customer',1:'High_Volume_Customer'}
Data['Cluster_Total']=Data['Cluster'].map(Cluster_Total)
print(Data[['Cust_ID','Cluster','Cluster_Total']].head())
```
